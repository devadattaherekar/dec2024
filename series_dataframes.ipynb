```python
#Series object of dtype object not str!
players=["virat","rohit","bumrah","jadeja"]
print(pd.Series(players))
players=[10,True,68.7] # heterogeneous elements in collection 
print(pd.Series(players)) #Series object of dtype object
```

```python
#Index need not be int64
dream11={"sachin":10,"rohit":45,"gayle":333}
print(pd.Series(dream11)) # dtype refers to values
```

```python
#Series object has an advantage over list that index can be any dtype
dream11={"sachin":10,"rohit":45,"gayle":333}
print(pd.Series(index=[1,1,2,2],data=[10,10,20,30])) # dtype refers to values
#Index need not be int64
```

```python
#Series object has an advantage over dict that index can be duplicates
test_dict = {'a': 10, 'b': 20, 'c': 30}
series = pd.Series(data=test_dict, index=['a', 'y', 'z'])
print(series)
```

```python
#Attributes give mostly information about object they don't modify object unlike methods
days=["sun","mon","tue","sun"]
d=pd.Series(days)
print(d) #values,index,dtype
print(d.values)
print(d.index)
print(d.dtype)
print("Name of Series is ",d.name)
d.name="Days of the Week"
print("Name of Series is ",d.name)
print(d.ndim) # will return 2 for dataframe
print(d.size) #return rows*col including blank rows or cols
print(d.shape) #returns tuple with (rows,col) 
print(d.is_unique)
```

```python
#methods perform operation
prices=[99,80,100,65,77]
s=pd.Series(prices)
print(s.sum())
print(s.subtract(5))
print(s.prod())
print(s.product())
print(s.divide(2))
print(s.mean())
print(s.median())
print(s.mode())
print(s.describe())
print(len(s))
print(s.count()) # count does not count Nan!
```

```python
#sort_index and sort_values creates new object 
#assign or use inplace parameter to modify Series 
csv_file=pd.read_csv('sample_datasets/artists.csv',usecols=[1])
ps=csv_file.squeeze()
#print(ps.sort_index(ascending=False))
#print(ps)
ps.sort_index(ascending=False,inplace=True)
print(ps.head(1))
ps=ps.sort_index(ascending=True) # Here inplace if used returns NoneType!
print(ps.head(1))
```

```python
#sort_values creates new object 
#assign or use inplace parameter to modify Series 
csv_file=pd.read_csv('sample_datasets/artists.csv',usecols=[1])
ps=csv_file.squeeze()
print(ps.sort_values(ascending=False))# does not modify Series object
print("Largest value is ",ps.sort_values(ascending=False).head(1)) # returns Series object which can be used to cascade
print(ps.head(3)) 
ps=ps.sort_values()#Here this creates a new object
print(ps.head(n=3))
ps.sort_values(ascending=False,inplace=True) # returns NoneType so dont assign to any reference!
print(ps.head(n=3))
#print(ps.set_axis(labels=list(range(1,308))))
```

```python
#using apply() method
def check(number):
    if number >=50:
        return "superb"
    elif number >= 0 and number <50:
        return "ok"
        
csv_file=pd.read_csv('sample_datasets/numbers.csv')
csv_file=csv_file.squeeze()
print(csv_file.tail())
print(csv_file.apply(lambda no:no*no).tail())
print(csv_file.apply(check).tail())
```
