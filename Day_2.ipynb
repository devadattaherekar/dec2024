```python
# basic_salary is input() by user
# HRA 40%basic_salary # DA 30%basic_salary 
# PF=3000 PT=2000
# gross_salary=basic_salary+HRA+DA
# net_salary=gross_salary-PF-PT
# use cstyle/str.format/fstring
# use proper conversion functions and follow PEP8 and PEP257 guidelines when your write your code!
```

```python
basic_sal=eval(input("Enter basic salary"))
HRA=.20*basic_sal
DA=30/100*basic_sal
PF=3000
PT=2000
gross_sal=basic_sal+HRA+DA+PF+PT
net_sal=gross_sal-PF-PT
print(f"Net salary = {net_sal}\nHRA = {HRA}\nDA = {DA} \n")
print("Net salary = {0}\nHRA = {1}\nDA = {2} \n".format(net_sal,HRA,DA))
print("Net salary = %.2f\nHRA = %.2f\nDA = %.2f \n"%(net_sal,HRA,DA))
```
